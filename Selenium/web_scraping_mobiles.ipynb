```python
import csv
from selenium import webdriver

max_page_num = 5
page_num = 1

with open("mobiles.csv", "w") as f:
    f.write("Title, Price \n")
    
driver = webdriver.Chrome()

for i in range(1,max_page_num):            
    page_num += 1    
    url = "https://propakistani.pk/price/brand/xiaomi/page/" + str(page_num) + "/"    
    
    driver.get(url)
    
    title = driver.find_elements_by_xpath('//div[@class="title"]')
    price = driver.find_elements_by_xpath('//h6[@class="mb-1 text-warning medium"]')
    
    num_page_items = len(title)
    with open("mobiles.csv", "a") as f:
        for i in range(num_page_items):
            f.write(title[i].text + "," + price[i].text + "\n")
    
driver.close()
```
